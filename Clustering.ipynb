# Clustering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D

mall_data_path = r"C:\Users\DELL\Downloads\Mall_Customers.csv"
data = pd.read_csv(mall_data_path)
print(data.head())
```

```python
# missing values
print(data.isnull().sum())
```

```python
# Select features for clustering
# For 2D clustering
X_2D = data[['Annual Income (k$)', 'Spending Score (1-100)']]

# For 3D clustering
X_3D = data[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']]

#  K-Means clustering with a chosen number of clusters
n_clusters = 5
kmeans_2D = KMeans(n_clusters=n_clusters, random_state=42)
kmeans_2D.fit(X_2D)

kmeans_3D = KMeans(n_clusters=n_clusters, random_state=42)
kmeans_3D.fit(X_3D)

data['Cluster_2D'] = kmeans_2D.labels_
data['Cluster_3D'] = kmeans_3D.labels_
```

# 2D Cluster Plot 

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_2D['Annual Income (k$)'], X_2D['Spending Score (1-100)'], c=data['Cluster_2D'], cmap='viridis')
plt.title('2D K-Means Clustering')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

# 3D Cluster Plot

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_3D['Age'], X_3D['Annual Income (k$)'], X_3D['Spending Score (1-100)'], c=data['Cluster_3D'], cmap='viridis')
ax.set_title('3D K-Means Clustering')
ax.set_xlabel('Age')
ax.set_ylabel('Annual Income (k$)')
ax.set_zlabel('Spending Score (1-100)')
plt.show()
```
